```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.plot()
```

```python
x = [50,100,150,200,250,300]
y = [2,10,25,70,80,120]
plt.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
```

```python
fig = plt.figure()
ax = fig.add_axes((1,1,1,1))
ax.plot(x,y)
plt.show()
```

```python
fig,ax = plt.subplots(figsize = (5,5))
ax.plot(x,y)
plt.plot()
```

```python
fig,ax = plt.subplots(figsize = (7,7))
ax.set(title = 'Simple Plot', xlabel = '50s', ylabel ='20s')
ax.plot(x,y)
plt.plot()
```

```python
fig,ax = plt.subplots()
ax.set(title  = 'Simple Plot')
ax.plot(x,y)
plt.plot()
```

```python
x = np.linspace(0,10,100)
x[:10]
fig,ax = plt.subplots()
ax.plot(x,x**4)
```

```python
np.random.seed(0)
stud_names = ['Jane','Joe','Beck','Tom','Sam',
'Eva','Samuel','Jack','Dana','Ester',
'Carla','Steve','Fallon','Liam','Culhane',
'Candance','Ana','Mari','Steffi','Adam']
stud_mark = np.random.randint(25,100,20)
fig,ax = plt.subplots()
ax.plot(stud_names,stud_mark,markerfacecolor ='red',marker = 'o',linestyle = 'dashed',scalex = True,scaley =True, linewidth = '2')
plt.xticks(rotation = 90)
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
np.random.seed(0)
N = 100
x = np.random.randint(1,100,N)
y = np.random.randint(1,100,N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))
fig,ax = plt.subplots()
ax.scatter(x,y,s = area,c = colors)
```

```python
food_sales = {'Chapathi sales': 150.00,
'Dosa sales': 225.00,
'Puri Sales': 75.00}
fig,ax = plt.subplots()
ax.bar(food_sales.keys(),height=food_sales.values())
ax.set(title = 'food_sales',xlabel = 'Food Items', ylabel = 'No.Units')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.barh(food_sales.keys(),width=food_sales.values())
ax.set(title ='Food Sales',xlabel = 'No.Units',ylabel = 'Food Items')
plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x)
```

```python
fig,ax = plt.subplots()
ax.scatter(x,y)
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2)
ax1.plot(x,y)
ax2.scatter(x,np.sin(x))
ax3.bar(food_sales.keys(),food_sales.values())
ax4.hist(np.random.randn(10000))
```

```python
time_series = pd.Series(np.random.randn(1000),
index = pd.date_range('1/1/2022',periods = 1000))
time_series = time_series.cumsum()
fig,ax = plt.subplots()
time_series.plot()
```
