```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
df.head()
sns.scatterplot(x='total_bill',y='tip',data = df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data = df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data = df,style = 'sex', kind = 'scatter')
plt.show()
```

```python
sns.relplot(x='total_bill',y = 'tip',hue = 'sex',col = 'time',data =df)
plt.show()
```

```python
df_flights = sns.load_dataset('flights')
df_flights.head()
dots = sns.relplot(x='year',y='passengers',data = df_flights,kind = 'scatter',hue = 'month')
plt.show(dots)
line = sns.relplot(x='year',y='passengers',data = df_flights,kind = 'line',hue = 'month')
plt.show(line)
```

```python
Strip = sns.catplot(x='sex',y='total_bill',data = df,kind = 'strip',hue = 'time')
plt.show(Strip)
sns.catplot(x='sex',y='total_bill',data = df,kind = 'strip',hue = 'time',jitter = 0.05)
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill', data = df , kind = 'swarm',hue = 'time')
plt.show()
```

```python
sns.catplot(x = 'sex',y = 'total_bill',data = df , kind = 'box')
plt.show()
```

```python
sns.catplot(x = 'sex',y = 'total_bill',data = df , kind = 'box',hue ='smoker')
plt.show()
```

```python
sns.catplot(x = 'sex',y = 'total_bill',data = df , kind = 'boxen',hue ='smoker')
plt.show()
```

```python
df_iris = sns.load_dataset('iris')
df_iris.head()
sns.catplot(x='species',y='petal_length',data = df_iris, kind = 'violin')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
sns.catplot(x='sex',y='survived',data = titanic,hue ='class' ,kind = 'bar')
plt.show()
```

```python
sns.catplot(x='deck',kind = 'count' , palette='ch:.25',data=titanic)
plt.show()
```

```python
sns.catplot(x='sex',y='survived',kind = 'point',data = titanic,hue = 'class')
plt.show()
```

```python
sns.catplot(x='sex',y='age',kind = 'bar',data = titanic,hue = 'class')
```

```python
data = np.random.multivariate_normal([0,0],[[5,2],[2,2]],size = 2000)
data_df = pd.DataFrame(data,columns=['x','y'])
for col in 'xy':
  plt.hist(data_df[col],alpha = 0.5)
```

```python
for col in 'xy':
   sns.kdeplot(data_df[col],shade = True)
```

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = 100*tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row='sex', col='time', margin_titles=True)
grid.map(plt.hist, 'tip_pct', bins=np.linspace(0, 40, 15))
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')
sns.heatmap(flights_pivot)
plt.show()
```

```python
sns.heatmap(flights_pivot, annot = True, fmt = 'd')
plt.show()
```
